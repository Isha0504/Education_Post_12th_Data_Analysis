* OUTLIERS

```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range= Q1-(1.5 * IQR)
    upper_range= Q3+(1.5 * IQR)
    return lower_range, upper_range

for i in df.columns:
    if df[i].dtype !='object':
        lr,ur=remove_outlier(df[i])
        df[i]=np.where(df[i]>ur,ur,df[i])
        df[i]=np.where(df[i]<lr,lr,df[i])
```

Insight - The outliers in the dataset are removed using Inter-Quartile Range method by calculating Q1(lower bound) and Q3(upper bound). The values lesser than the 'lower bound' and the valuer higher than the 'upper bound' are replaced with lower bound value and upper bound value respectively.

9. OVERALL INSIGHTS

	Which factors (applications, acceptance rate, enrollment, academic excellence, costs, faculty qualifications, student/faculty ratio, alumni donations, expenditures) are most strongly associated with higher graduation rates?

```python
numeric_df = df.select_dtypes(include=['number'])

# Compute the correlation matrix
correlation_matrix = numeric_df.corr()

# Factors of interest
factors_of_interest = ['Apps', 'Accept', 'Enroll', 'Top10perc', 'Top25perc',
                       'Outstate', 'Room.Board', 'Books', 'Personal', 'PhD',
                       'Terminal', 'S.F.Ratio', 'perc.alumni', 'Expend']

# Ensure factors of interest are in the DataFrame
factors_of_interest = [factor for factor in factors_of_interest if factor in numeric_df.columns]

# Identify the correlations of these factors with graduation rate
correlation_with_gradrate = correlation_matrix.loc[factors_of_interest, 'Grad.Rate']
correlation_with_gradrate = correlation_with_gradrate.sort_values(ascending=False)
print("Factors most strongly correlated with graduation rates:")
print(correlation_with_gradrate)

# Visualizing the correlations
plt.figure(figsize=(12, 8))
sns.barplot(x=correlation_with_gradrate.values, y=correlation_with_gradrate.index, palette='viridis')
plt.title('Correlation of Various Factors with Graduation Rate')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Factors')
plt.grid(True)
plt.show()

# Visualize top 3 factors with scatter plots
top_factors = correlation_with_gradrate.index[:3]
for factor in top_factors:
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x=factor, y='Grad.Rate')
    plt.title(f'{factor} vs Graduation Rate')
    plt.xlabel(factor)
    plt.ylabel('Graduation Rate')
    plt.grid(True)
    plt.show()
```

	What recommendations can be made to colleges to improve their graduation rates based on the data analysis?

```python
strongest_factors = correlation_with_gradrate.index[:5]  # Get top 5 factors

# Recommendations
recommendations = {
    'Top10perc': "Increase the intake of students from the top 10% of their higher secondary class.",
    'Top25perc': "Increase the intake of students from the top 25% of their higher secondary class.",
    'perc.alumni': "Enhance alumni engagement and encourage alumni donations to create a supportive network.",
    'PhD': "Increase the percentage of faculty with Ph.D.s to improve academic quality and mentorship.",
    'Terminal': "Increase the percentage of faculty with terminal degrees to ensure high academic standards.",
    'S.F.Ratio': "Reduce the student-faculty ratio to ensure more personalized attention and support for students.",
    'Expend': "Increase instructional expenditure per student to improve resources and academic support services."
}

print("\nRecommendations to Improve Graduation Rates:")
for factor in strongest_factors:
    if factor in recommendations:
        print(f"- {recommendations[factor]}")

```
